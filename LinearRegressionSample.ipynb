```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns
```

```python
#import data
salaryData = pd.read_csv('SalaryData.csv')

x = salaryData.iloc[:, :1].values
y = salaryData.iloc[:, 1:].values
```

```python
#split data set into testing and training data
from sklearn.model_selection import train_test_split as tts 
x_train, x_test, y_train, y_test = tts(x, y , test_size = 0.3, random_state = 0)
```

```python
#
# Linear Regression
#

# Step 1: Fit linear regression to training data set
from sklearn.linear_model import LinearRegression 
lr = LinearRegression()
lr.fit(x_train, y_train)

#Step 2: Prediction of test results

y_pred = lr.predict(x_train)

plt.scatter(x_train,y_train, color = 'green')
plt.plot(x_train, y_pred, color = 'red')
plt.show()
```

```python
plt.scatter(x_test,y_test, color = 'green')
plt.plot(x_train, y_pred, color = 'red')
plt.show()
```
